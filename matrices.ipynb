```python
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Phylo.TreeConstruction import DistanceCalculator
import pandas as pd

alignment_path = "/content/max.afa"
alignment = AlignIO.read(alignment_path, "fasta")
cleaned_seqs = []
for record in alignment:
    ungapped = [aa for aa in record.seq if aa not in "-X"]
    if len(ungapped) / len(record.seq) > 0.3:
        cleaned_seqs.append(record)

cleaned_alignment = MultipleSeqAlignment(cleaned_seqs)
calculator = DistanceCalculator('identity')
dm = calculator.get_distance(cleaned_alignment)
df = pd.DataFrame(dm.matrix, index=dm.names, columns=dm.names)
output_path = "/content/distance_matrix_cleaned.csv"
df.to_csv(output_path)

df.head()
```

```python
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform
import pandas as pd
df = pd.read_csv("/content/distance_matrix_cleaned.csv", index_col=0)
df_full = df.combine_first(df.T)
df_filled = df_full.fillna(df_full.max().max() + 0.1)
condensed_dist = squareform(df_filled.values)


Z = linkage(condensed_dist, method='average')  # 'ward', 'single', 'complete', etc.
k = 4
clusters = fcluster(Z, k, criterion='maxclust')
cluster_df = pd.DataFrame({
    'Sequence': df_filled.index,
    'Cluster': clusters
})
cluster_df.to_csv("/content/sequence_clusters.csv", index=False)

cluster_df.head(10)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform
import re

path = "/content/distance_matrix_cleaned.csv"
df = pd.read_csv(path, index_col=0)
df_sym = df.combine_first(df.T).fillna(df.max().max() + 0.1)
condensed = squareform(df_sym.values)
Z = linkage(condensed, method='average')
k = 4
clusters = fcluster(Z, k, criterion='maxclust')

def extract_species_code(seq):
    match = re.search(r"_([A-Z0-9]+)$", seq)
    return match.group(1) if match else "UNKNOWN"

cluster_df = pd.DataFrame({
    "Sequence": df_sym.index,
    "Cluster": clusters
})
cluster_df["Species_Code"] = cluster_df["Sequence"].apply(extract_species_code)
cluster_df.to_csv("/content/annotated_sequence_clusters.csv", index=False)
cluster_df = cluster_df.set_index("Sequence").loc[df_sym.index]
palette = sns.color_palette("hsv", k)
cluster_colors = [palette[c - 1] for c in clusters]

sns.clustermap(df_sym, cmap="viridis", figsize=(20, 20),
               row_colors=cluster_colors, col_colors=cluster_colors,
               metric="euclidean", method="average")
plt.suptitle("Clustered Heatmap with Group Coloring", y=1.02)
plt.show()

for i in range(1, k + 1):
    seqs = cluster_df[cluster_df["Cluster"] == i].index
    sub_df = df_sym.loc[seqs, seqs]
    plt.figure(figsize=(8, 6))
    sns.heatmap(sub_df, cmap="viridis", square=True)
    plt.title(f"Cluster {i}: {len(seqs)} sequences")
    plt.xticks([], [])
    plt.yticks([], [])
    plt.tight_layout()
    plt.show()

grouped = {i: df_sym.loc[cluster_df[cluster_df["Cluster"] == i].index] for i in range(1, k + 1)}
avg_matrix = np.zeros((k, k))
for i in range(1, k + 1):
    for j in range(1, k + 1):
        sub_df = grouped[i][grouped[j].index]
        avg_matrix[i - 1, j - 1] = sub_df.values.mean()

avg_df = pd.DataFrame(avg_matrix, index=[f"Cluster {i}" for i in range(1, k + 1)],
                      columns=[f"Cluster {i}" for i in range(1, k + 1)])
print("Average Distance Between Clusters:")
display(avg_df)

cluster_df.reset_index().to_csv("/content/sequence_clusters.csv", index=False)
avg_df.to_csv("/content/cluster_distance_matrix.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/annotated_sequence_clusters.csv")

def fill_species_name(code):
    if code == "PLAF7":
        return "Plasmodium falciparum (3D7)"
    elif code == "PLAF4":
        return "Plasmodium falciparum (FCR-3)"
    elif code == "PLAF8":
        return "Plasmodium falciparum (Dd2)"
    elif code == "PLAFA":
        return "Plasmodium falciparum (ATCC 30932)"
    elif code.startswith("PLAF"):
        return "Plasmodium falciparum (other strains)"
    elif code == "PLAGA":
        return "Plasmodium gallinaceum"
    elif code == "PLARL":
        return "Plasmodium reichenowi"
    elif code == "PLABA":
        return "Plasmodium berghei"
    elif code == "PLARE":
        return "Plasmodium relictum"
    elif code == "PLAMA":
        return "Plasmodium malariae"
    elif code == "PLAKN":
        return "Plasmodium knowlesi"
    elif code == "9APIC":
        return "Other Apicomplexa"
    else:
        return f"Plasmodium (code: {code})"


df["Species_Full"] = df["Species_Code"].apply(fill_species_name)
plot_data = df.groupby(["Cluster", "Species_Full"]).size().reset_index(name="Count")
plt.figure(figsize=(20, 12))
sns.barplot(data=plot_data, x="Count", y="Species_Full", hue="Cluster", palette="tab10")
plt.title("Species Composition Across All Clusters")
plt.xlabel("Number of Sequences")
plt.ylabel("Species")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```
